# Data Ingestion

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

df= pd.read_csv("../data_raw/raw_StudentsPerformance.csv")
```

# Data Visualization

```python
# Gender differences in math vs reading (boxplots)
df.boxplot(column=['math score', 'reading score'], by='gender', figsize=(8, 6))
plt.suptitle('Math and Reading Scores vs Gender')
plt.savefig('../reports/Gender_Boxplots.png', dpi=300, bbox_inches='tight')

# Math Score vs Completed Test Prep
df.boxplot(column='math score', by='test preparation course', figsize=(8, 6))
plt.suptitle('Math Scores vs Test Preperation')
plt.title('')
plt.ylabel('Math Scores (0-100)%')
plt.xlabel('Test Preperation Course')
plt.savefig('../reports/Test_Prep_Impact.png', dpi=300, bbox_inches='tight')

# Does lunch type affect scores
df.groupby(by='lunch').mean(numeric_only=True).plot.bar(figsize=(8, 6))
plt.title('Average Test Scores vs Lunch Type')
plt.ylabel('Average Scores (0-100)%')
plt.xlabel('Lunch Type')
plt.savefig('../reports/Lunch.png', dpi=300, bbox_inches='tight')

# Subject Correlation (correlation heatmap for math, reading, and writing)
plt.figure(figsize=(8, 6))
df.corr(numeric_only=True)
sns.heatmap(df.corr(numeric_only=True))
plt.title('Subject Correlation')
plt.savefig('../reports/Subject_Correlation_Heatmap.png', dpi=300, bbox_inches='tight')

# Math vs Reading with trend lines by test prep
plt.figure(figsize=(8, 6))
sns.lmplot(data=df, x='math score', y='reading score', hue='test preparation course')
plt.title('Reading vs Math')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.savefig('../reports/Math_and_Reading_Association.png', dpi=300, bbox_inches='tight')
```
